```python
from train import pts_preds, pts_labels, assist_preds, assist_labels, reb_preds, reb_labels, epoch_enum, validation_errors
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
from sklearn.metrics import root_mean_squared_log_error, r2_score, d2_absolute_error_score, d2_tweedie_score
from sklearn.metrics import PredictionErrorDisplay
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Root mean squared error evaluation metric for predictions of points, assists, rebounds
# We can interpret this error as: [rmse_pts, rmse_assists, rmse_rebs] points, assists, rebounds respectively
rmse_pts = root_mean_squared_error(pts_labels, pts_preds)
print("RMSE for points: " + str(rmse_pts))
rmse_assists = root_mean_squared_error(assist_labels, assist_preds)
print("RMSE for assists: " + str(rmse_assists))
rmse_rebs = root_mean_squared_error(reb_labels, reb_preds)
print("RMSE for rebounds: " + str(rmse_rebs))
```

```python
# Mean Absolute Error evaluation metric for predictions of points, assists, rebounds
# We can interpret this error as: [mae_pts, mae_assists, mae_rebs] points, assists, rebounds respectively
mae_pts = mean_absolute_error(pts_labels, pts_preds)
print("MAE for points: " + str(mae_pts))
mae_assists = mean_absolute_error(assist_labels, assist_preds)
print("MAE for assists: " + str(mae_assists))
mae_rebs = mean_absolute_error(reb_labels, reb_preds)
print("MAE for rebounds: " + str(mae_rebs))
```

```python
# Root Mean Squared Log Error evaluation metric for predictions of points, assists, rebounds
# We can interpret this error as: [rmsle_pts, rmsle_assists, rmsle_rebs] points, assists, rebounds respectively
rmsle_pts = root_mean_squared_log_error(pts_labels, pts_preds)
print("RMSLE for points: " + str(rmsle_pts))
rmsle_assists = root_mean_squared_log_error(assist_labels, assist_preds)
print("RMSLE for assists: " + str(rmsle_assists))
rmsle_rebs = root_mean_squared_log_error(reb_labels, reb_preds)
print("RMSLE for rebounds: " + str(rmsle_rebs))
```

```python
#R^2 Score and D^2 Absolute Error Score Evaluation Metric 
r2_pts = r2_score(pts_labels, pts_preds)
print("R^2 score for points: " + str(r2_pts))
r2_assists = r2_score(assist_labels, assist_preds)
print("R^2 score for assists: " + str(r2_assists))
r2_rebs = r2_score(reb_labels, reb_preds)
print("R^2 score for rebounds: " + str(r2_rebs))


d2_pts = d2_absolute_error_score(pts_labels, pts_preds)
print("D^2 Absolute Error score for points: " + str(d2_pts))
d2_assists = d2_absolute_error_score(assist_labels, assist_preds)
print("D^2 Absolute Error score for assists: " + str(d2_assists))
d2_rebs = d2_absolute_error_score(reb_labels, reb_preds)
print("D^2 Absolute Error score for rebounds: " + str(d2_rebs))
```

```python
# Actual vs. Predicted Point Values Per Game
x_pts, y_pts = np.array(pts_preds), np.array(pts_labels)
m, b = np.polyfit(x_pts, y_pts, 1)
plt.scatter(x_pts, y_pts)
plt.plot(x_pts, ((m*x_pts)+b))
plt.ylabel("Actual Points Per Game")
plt.xlabel("Predicted Points Per Game")
plt.title("Actual vs. Predicted Points Per Game")
plt.show()

#Actual vs. Predicted Assists Per Game
x_assists, y_assists = np.array(assist_preds), np.array(assist_labels)
m1, b1 = np.polyfit(x_assists, y_assists, 1)
plt.scatter(x_assists, y_assists)
plt.plot(x_assists, ((m1*x_assists)+b1))
plt.ylabel("Actual Assists Per Game")
plt.xlabel("Predicted Assists Per Game")
plt.title("Actual vs. Predicted Assists Per Game")
plt.show()

#Actual vs. Predicted Rebounds Per Game
x_rebs, y_rebs = np.array(reb_preds), np.array(reb_labels)
m2, b2 = np.polyfit(x_rebs, y_rebs, 1)
plt.scatter(x_rebs, y_rebs)
plt.plot(x_rebs, ((m2*x_rebs)+b2))
plt.ylabel("Actual Rebounds Per Game")
plt.xlabel("Predicted Rebounds Per Game")
plt.title("Actual vs. Predicted Rebounds Per Game")
plt.show()
```

```python
#Graphing the predicted Error
pts_display = PredictionErrorDisplay(y_true=y_pts, y_pred=x_pts)
pts_display.plot()
plt.title("Residuals vs. Predicted Points Per Game")
plt.show()

assists_display = PredictionErrorDisplay(y_true=y_assists, y_pred=x_assists)
assists_display.plot()
plt.title("Residuals vs. Predicted Assists Per Game")
plt.show()

rebs_display = PredictionErrorDisplay(y_true=y_rebs, y_pred=x_rebs)
rebs_display.plot()
plt.title("Residuals vs. Predicted Rebounds Per Game")
plt.show()
```
